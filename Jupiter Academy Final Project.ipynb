### Handling missing values

#### 1. Handle missing values in 'GENDER'

```python
# Check for the value counts of unique values in 'GENDER'
df['GENDER'].value_counts()
```

```python
# Replace missing values with 'M' since there are less number of males
df['GENDER'] = df['GENDER'].fillna('M')

# Check for the unique values in 'GENDER'
df['GENDER'].unique()
```

#### 2. Handle missing values in 'Annual_income'

```python
# Fill missing values with mean of df['Annual_income']
df['Annual_income'] = df['Annual_income'].fillna(df['Annual_income'].mean())
```

#### 3. Handle missing values in 'Birthday_count'

```python
# Fill missing values with mean of df['Birthday_count']
df['Birthday_count'] = df['Birthday_count'].fillna(df['Birthday_count'].mean())
```

#### 4. Handle missing values in 'Type_Occupation'

```python
# Replace missing values with 'Others'
df['Type_Occupation'] = df['Type_Occupation'].fillna('Others')

# Check for unique values in df['Type_Occupation']
df['Type_Occupation'].unique()
```

```python
# Check to see if all missing values have be handled
df.isnull().sum()
```

```python
# Final view
for columns in df:
    unique_values = df[columns].unique()
    print(f'{columns}: {unique_values}')
```

```python
# Drop 'Ind_ID' feature
# Drop 'mobile_phone' feature because all the values are 1.
df.drop(['Ind_ID','Mobile_phone'], axis=1, inplace=True)
df
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

### Building Machine Learning Models

```python
# copy df into another dataframe
df2 = df.copy()
```

#### Skewness

```python
# Check for skewness
df2.skew()
```

```python
# Reduce skewness values for those greater than 1
num_columns = ['CHILDREN', 'Annual_income', 'Work_Phone', 'EMAIL_ID']

from scipy.stats import skew
for cols in num_columns:
    if df2[cols].skew()>1:
        df2[cols]= np.log1p(df2[cols])
df2.skew()
```

#### Encoding

```python
# Convert categorical variables in numerical variables
df2 = pd.get_dummies(df2)
df2
```

```python
# Split dataset into x and y
x = df2.drop(['label'], axis=1)
y = df2[['label']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=42)
```

```python
# Using Support Vector Machine
from sklearn.svm import SVC

model = SVC()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
# Using Random Forest Classifier to check for a better accuracy score
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
model2.score(x_test, y_test)
```

#### Feature Scaling

```python
# Scale Features to improve accuracy scores of models
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()

scaled_features= [features for features in x.columns]
scaled_data= scaler.fit_transform(x)

x= pd.DataFrame(data=scaled_data, columns=scaled_features)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=42)

model = SVC()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
model2.score(x_test, y_test)
```

```python
# Using model (Support Vector Machine)
y_pred = model.predict(x_test)
y_pred
```
