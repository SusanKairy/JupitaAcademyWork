### 1. Handling missing data

### Handling missing values in LOAN DATE END column

```python
# Since Loan Date End values are repeated in Contract column; drop the Loan Date End column

df.drop('Loan Date End', axis = 1, inplace = True)
```

### Handling missing values in HITS column

```python
df['Hits'].mode()
```

```python
df['Hits'] = df['Hits'].fillna(0)
```

```python
# Check again for columns with missing values
missing = df.isnull().sum()
missing = missing[missing>0]
missing
```
