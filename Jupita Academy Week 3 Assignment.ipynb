```python
# 2. Using these data points create a bar plotcategories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

data = {'A':25, 'B':40, 'C':15, 'D':30}
        
Categories = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (8, 5))
 
# creating the bar plot
plt.bar(Categories, values, width = 0.5)
 
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
# 3. Create a histogram using these data pointsdata = [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]

x=[32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60] 
sns.histplot(x, bins=10)

plt.title('Histogram')
plt.xlabel('Points')
plt.ylabel('Count')

plt.show()
```
