### Importing necessary Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# 1. Using these data points create a line plot

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 7]

plt.plot(x,y);
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('Line Plot')
plt.show()
```

```python
# 2. Using these data points create a bar plot

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

data = {'A':25, 'B':40, 'C':15, 'D':30}
        
Categories = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (8, 5))
 
# creating the bar plot
plt.bar(Categories, values, color ='purple', width = 0.5)
 
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
# 2. Using these data points create a bar plot

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

data = {'Categories':['A', 'B', 'C', 'D'],
       'Values': [25, 40, 15, 30]}
df = pd.DataFrame(data)

sns.barplot(x= 'Categories', y= 'Values', data=df);
```

```python
# 3. Create a histogram using these data points 

data = [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]

sns.histplot(data, bins=10)

plt.title('Histogram')
plt.xlabel('Points')
plt.ylabel('Count')

plt.show()
```

```python
# 4. Using these data points, create a scatterplot

x = [3, 5, 7, 10, 12, 15]
y = [25, 40, 30, 50, 35, 60]

sns.scatterplot(x = [3, 5, 7, 10, 12, 15], y = [25, 40, 30, 50, 35, 60])

plt.title('Scatterplot')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')

plt.show()
```

```python
# 5. Using these data points, create a pie chart 

labels = ['Apples', 'Bananas', 'Oranges', 'Grapes'] 
sizes = [30, 25, 20, 15]

y = sizes

plt.pie(y, labels= labels, autopct='%.0f%%')
plt.show()
```

```python
# 6. Create a box plot using these data points

data= [45, 60, 75,300, 80,350, 90, 95, 100, 110, 120,250,200]
sns.boxplot(data);
```
